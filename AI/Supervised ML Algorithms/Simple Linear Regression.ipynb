```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
warnings.filterwarnings('ignore')
fuel_df = pd.read_csv('fuel_efficiency_data.csv')

# print(fuel_df.head())
# print(fuel_df.describe())
```

```python
engine_size = fuel_df[['Engine_Size_Liters']]
fuel_efficieny = fuel_df['Fuel_Efficiency_MPG']

plt.title("Fuel Efficieny Dataset")
plt.xlabel('Engine Size (L)')
plt.ylabel('Fuel Efficiency (MPG)')

plt.scatter(engine_size, fuel_efficieny, marker='o')

x_train,x_test,y_train,y_test = train_test_split(engine_size,fuel_efficieny,test_size=0.30, random_state=75)

lr_model = LinearRegression()

lr_model.fit(x_train,y_train)
```

```python
y_hat = lr_model.predict(x_test)
y_hat
```

```python
plt.scatter(x_train,y_train,color="blue",label="Training Set")
plt.scatter(x_test,y_test,color="green",label="Testing Set")
plt.plot(x_test,y_hat,color="salmon", label="Regression Line?")

plt.legend()
```

```python
new_entry=input("Please enter new engine size")
new_entry=np.array([[float(new_entry)]])

new_prediction=lr_model.predict(new_entry)

print(f"The fuel eff for the new engine size is {new_prediction}")
```

```python
print(f"the m for this model is {lr_model.coef_}")
print(f"the b for this model is {lr_model.intercept_}")
```

```python
mse = mean_squared_error(y_hat, y_test)
mse
```
